```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
df = pd.merge(ratings, movies, on='movieId')
```

```python
print("Preview of merged dataset:")
print(df.head())
print("\nDataset info:")
print(df.info())
print("\nStatistical summary:")
print(df.describe())
```

```python
ratings_per_movie = df.groupby('title')['rating'].count().sort_values(ascending=False)
print("Top 10 most rated movies:")
print(ratings_per_movie.head(10))
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
popular_movies = ratings_per_movie[ratings_per_movie >= 10].index
filtered_df = df[df['title'].isin(popular_movies)]
print("Filtered dataset shape:", filtered_df.shape)
```

```python
user_movie_matrix = filtered_df.pivot_table(index='userId', columns='title', values='rating')
print("User-Movie matrix shape:", user_movie_matrix.shape)
```

```python
genre_dummies = movies['genres'].str.get_dummies(sep='|')
movies_with_genres = pd.concat([movies[['movieId']], genre_dummies], axis=1)
```

```python
df_with_genres = pd.merge(df, movies_with_genres, on='movieId')
user_stats = df_with_genres.groupby('userId')['rating'].agg(['mean', 'std', 'count']).reset_index()
user_stats.columns = ['userId', 'user_avg_rating', 'user_std_rating', 'user_rating_count']
```

```python
user_stats['user_std_rating'] = user_stats['user_std_rating'].fillna(0)
df_features = pd.merge(df_with_genres, user_stats, on='userId')

```

```python
df_features = pd.merge(df_with_genres, user_stats, on='userId')
feature_columns = ['user_avg_rating', 'user_std_rating', 'user_rating_count'] + list(genre_dummies.columns)
feature_matrix = df_features[feature_columns + ['rating']]
```

```python
from sklearn.preprocessing import MinMaxScaler

```

```python
scaler = MinMaxScaler()
numeric_cols = ['user_avg_rating', 'user_std_rating', 'user_rating_count']
feature_matrix.loc[:, numeric_cols] = scaler.fit_transform(feature_matrix[numeric_cols])
```

```python
feature_matrix.to_csv('feature_matrix.csv', index=False)
print("Feature matrix saved to 'feature_matrix.csv'")
```
